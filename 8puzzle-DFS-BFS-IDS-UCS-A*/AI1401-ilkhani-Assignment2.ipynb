## general functions:

```python
from copy import deepcopy
from queue import PriorityQueue

class Node():
    def __init__(self, state, parent = None, depth = 0):
        self.parent = parent
        self.state = state
        self.child = []
        self.depth = depth
        self.f = None

    def __gt__(self, other):
        if(self.f>other.f):
            return True
        else:
            return False

    def __lt__(self, other):
        if(self.f<other.f):
            return True
        else:
            return False

    def __eq__(self, other):
        if(self.f == other.f):
            return True
        else:
            return False
    
    def __ge__(self, other):
        if(self.f>=other.f):
            return True
        else:
            return False

    def __le__(self, other):
        if(self.f<=other.f):
            return True
        else:
            return False

def biuld_child_up(state, i,j):
    s = deepcopy(state)
    s[i][j] = s[i-1][j]
    s[i-1][j] = 0
    return s

def biuld_child_down(state, i,j):
    s = deepcopy(state)
    s[i][j] = s[i+1][j]
    s[i+1][j] = 0
    return s

def biuld_child_left(state, i,j):
    s = deepcopy(state)
    s[i][j] = s[i][j-1]
    s[i][j-1] = 0
    return s

def biuld_child_right(state, i,j):
    s = deepcopy(state)
    s[i][j] = s[i][j+1]
    s[i][j+1] = 0
    return s


def remove_front(qu):
    return qu[0], qu[1:]
    
def build_children(node:Node):
    state = node.state
    children = []
    i,j = find_val(state, 0) # location of zero

    if i != 0:
        children.append(Node(biuld_child_up(state, i, j),node, node.depth+1))
    if i != 2:
        children.append(Node(biuld_child_down(state, i, j), node, node.depth+1))
    if j != 0:
        children.append(Node(biuld_child_left(state, i, j), node, node.depth+1))
    if j != 2:
        children.append(Node(biuld_child_right(state, i, j), node, node.depth+1))
    return children



def general_search(problem, queuingFn, statesQueue: queue.Queue) -> Node:
    global NUMBER_OF_NODES
    NUMBER_OF_NODES = 0
    visitedStates = set()
    statesQueue.put_nowait(Node(problem))
    while True:
        if statesQueue.empty():
            return 'failure'
        else:
            headState = statesQueue.get_nowait()
            
            visitedStates.add(tuple(map(tuple, headState.state)))
            if headState.state == GOAL_STATE:
                return headState
            else:
                validChildren = [child for child in build_children(headState) if tuple(map(tuple, child.state)) not in visitedStates]
                NUMBER_OF_NODES += len(validChildren)
                queuingFn(statesQueue, validChildren)


def find_val(state:list[list], val):
    for i in range(3):
        for j in range(3):
            if state[i][j] == val:
                return i,j

def heuristic(node:Node):
    h = 0 
    for iState in range(3):
        for jState in range(3):
            val = node.state[iState][jState]
            iTrue, jTrue = find_val(GOAL_STATE, val)
            return abs(iState-iTrue) + abs(jState - jTrue)
```

## BFS:

```python

def queue_fn_BFS(statesQu:queue.Queue, newStates:list[Node]):
    for s in newStates:
        statesQu.put_nowait(s)
```

## DFS:

```python
def queue_fn_DFS(statesQu:queue.LifoQueue, newStates:list[Node],maxDepth=np.Inf):
    if newStates != []:
        if newStates[0].depth > maxDepth:
            return
        for s in newStates:
            statesQu.put_nowait(s)
```

## IDS

```python
def IDS(problem):
    local_number_of_nodes = 0
    global NUMBER_OF_NODES
    depth = 1
    result = None
    while True:
        local_number_of_nodes += NUMBER_OF_NODES
        result = general_search(problem, lambda stateQu, newStates: queue_fn_DFS(stateQu, newStates, depth ), queue.Queue())
        if not isinstance(result, str):
            break
        depth += 1
    NUMBER_OF_NODES = local_number_of_nodes
    return result
            
```

```python
def queue_fn_AStar(stateQu:queue.PriorityQueue, newStates:list[Node]):
    if newStates != []:
        g = newStates[0].depth
        for s in newStates:
            s.f = heuristic(s) + g
            stateQu.put_nowait(s)
```
